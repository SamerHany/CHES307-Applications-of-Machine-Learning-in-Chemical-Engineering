<h2>1. Convolutional Neural Networks (CNN)</h2>


### **Overview**
Convolutional Neural Networks (CNNs) have emerged as a cornerstone architecture within the realm of deep learning, particularly excelling in visual recognition tasks.  CNNs have proven to be highly effective in various image-related tasks, such as image classification, object detection, and image segmentation.

### **Convolution:**
Convolution lies at the heart of CNNs. It involves applying a set of learnable filters (also known as kernels or feature detectors) to the input image. These filters capture local patterns and spatial relationships between adjacent pixels. By convolving the filters with the input image, we obtain a set of feature maps that highlight important features at different scales and levels of abstraction. The convolutional operation is typically followed by an element-wise non-linear activation function, such as ReLU (Rectified Linear Unit), to introduce non-linearity and enhance the network's capacity to model complex relationships.

### **Pooling:**
Pooling is an essential operation in CNNs that contributes to their hierarchical and spatially invariant nature. It reduces the spatial dimensions of the feature maps while retaining their most salient information. The most common pooling technique is max pooling, which partitions each feature map into non-overlapping regions and outputs the maximum value within each region. Pooling helps in down-sampling the feature maps, reducing the computation and memory requirements, and providing some degree of translational invariance.

### **Architecture:**
- **Convolutional Layers:** Convolutional layers are the building blocks of CNNs. They consist of multiple filters that convolve with the input to produce feature maps. The number of filters determines the depth of the layer, while their sizes define the receptive fieldâ€”the area over which each filter operates. Convolutional layers are typically stacked to learn increasingly complex features from the input data.

- Pooling Layers: Pooling layers reduce the spatial dimensions of the feature maps obtained from convolutional layers. Max pooling is the most common pooling technique, but other variations like average pooling exist as well. Pooling helps in downsampling, reducing computational complexity, and capturing the most salient features while maintaining spatial invariance.

- **Fully Connected Layers:** Fully connected layers are typically placed at the end of the CNN architecture. They take the flattened feature maps from the preceding layers and connect every neuron to every neuron in the subsequent layers. These layers are responsible for high-level reasoning and making predictions based on the learned features.

- **Activation Functions:** Activation functions introduce non-linearity to the CNN architecture. Rectified Linear Unit (ReLU) is commonly used as the activation function in CNNs due to its simplicity and efficiency in training deep networks. Other activation functions like sigmoid and hyperbolic tangent (tanh) can also be used, depending on the requirements of the task.

- **Dropout:** Dropout is a regularization technique used to prevent overfitting in CNNs. It randomly sets a fraction of the input units to zero during training, which helps in reducing interdependencies among neurons and encourages the network to learn more robust features.

### **Advantages**
- **Hierarchical Feature Learning:** CNNs automatically learn hierarchical representations of data, allowing them to capture complex patterns and relationships at different levels of abstraction. This feature learning capability makes CNNs highly effective in tasks where recognizing and understanding hierarchical structures is crucial.

- **Spatial Invariance:** CNNs are inherently designed to be spatially invariant, meaning they can detect features regardless of their exact spatial location in the input image. This property makes CNNs robust to variations in translation, rotation, and scaling, making them suitable for tasks like object detection and image recognition.

- **Parameter Sharing:** CNNs exploit the concept of parameter sharing, where the same set of filters is applied to different spatial locations of the input. This significantly reduces the number of parameters and allows CNNs to efficiently learn from large-scale datasets.

<hr width="50%" style="margin-top: 2rem; margin-bottom: 2rem; margin-left:0">

### **Disadvantages**
- **High Computational Cost:** CNNs can be computationally expensive, especially when dealing with large-scale datasets and complex architectures. Training and evaluating CNNs often require significant computational resources, including high-performance GPUs.

- **Requirements for Large Datasets:** CNNs typically require a large amount of labeled training data to generalize well. Obtaining and annotating large datasets can be challenging and time-consuming, particularly for specialized domains.

<hr width="50%" style="margin-top: 2rem; margin-bottom: 2rem; margin-left:0">

### **Selection Guidelines**
CNNs are particularly well-suited for tasks involving image and spatial data analysis. Here are some scenarios where CNNs are commonly applied:

- **Image classification:** Assigning a label or category to an input image.
 
- **Object detection:** Identifying and localizing objects within an image.

- **Image segmentation:** Assigning a class label to each pixel in an image.

### **Example:** Automated Casting Defect Detection using Image Classification

### Problem Overview

In the manufacturing industry, maintaining high-quality products is crucial for profitability.  Manual inspection is a common method for detecting defects during the casting process (where molten metal is shaped using molds). However, manual inspection suffers from limitations such as:

* **Inconsistent Accuracy:**  Inspection accuracy can vary significantly depending on the inspector's skill and experience.
* **Increased Labor Costs:**  Manual inspection requires substantial manpower, impacting production costs.

### Project Objective

The objective of this project is to develop a machine learning model that can automatically classify casting images as either "normal" or containing a "defect." This model will be trained on a labeled dataset of casting images and evaluated for its accuracy in identifying defects. 

**Success Criteria:**

* The model achieves an acceptable level of accuracy (e.g., above 90%) in classifying casting images as "normal" or "defective."
* The model demonstrates robustness to variations in lighting conditions that might be encountered during real-world deployment.

### Data

This project will utilize a dataset containing images of impellers for submersible pumps, obtained during the casting process. 

* **Data Type:** Images
* **Labels:** Each image will be labeled as either "ok" (normal) or "def" (defect).
* **Acquisition Considerations:** The dataset was acquired under controlled lighting conditions to minimize variability.

```python
# import libraries
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torchvision
from torchvision.transforms import transforms
from sklearn.metrics import confusion_matrix, classification_report
```

### Data Augmentation

It's a set of techniques that artificially expands the training data by creating modified versions of existing images. Data augmentation allows you to virtually generate more images from the existing ones, enriching the training data and leading to several benefits:

* **Reduced Overfitting:**  With a limited dataset, a CNN can become overly reliant on specific features within the training images. This can lead to overfitting, where the model performs well on the training data but poorly on unseen data. Data augmentation helps prevent this by introducing variations that the model must learn to be robust against. 

* **Improved Generalizability:**  By encountering a wider range of image transformations during training, the CNN becomes more adept at recognizing objects regardless of minor variations in factors like rotation, lighting, or scale. This translates to better performance on real-world data that might exhibit such variations.

Data augmentation techniques for CNNs can be broadly categorized into two groups:

* **Geometric Transformations:** These techniques modify the spatial properties of the image, such as:
    * Rotation: Flipping or rotating the image to different angles. 
    * Scaling: Zooming in or out on the image.
    * Cropping: Taking random crops from the original image, forcing the CNN to focus on relevant features within a smaller region.

* **Color Space Augmentations:** These techniques modify the color channels of the image:
    * Brightness and Contrast adjustments: Simulating different lighting conditions.
    * Color Jittering: Randomly changing the brightness, saturation, and hue of the image to introduce variations.

By strategically applying data augmentation techniques, you can significantly enhance the performance of your CNNs, allowing them to learn more effectively from a limited dataset and ultimately leading to superior results in real-world applications.


```python
# specify the directory where the data is stored
train_dir = 'casting_data/train'
test_dir = 'casting_data/test'

# apply different data augmentation transformations to a single image and plot the results
img = torchvision.datasets.ImageFolder(root=test_dir, transform=transforms.ToTensor())[-1][0]

plt.figure(figsize=(12, 12))

# original image
plt.subplot(2, 3, 1)
plt.title('Original')
plt.imshow(np.transpose(img, (1, 2, 0)))
plt.axis('off')

# rotation
plt.subplot(2, 3, 2)
plt.title('Rotation')
rot = transforms.RandomRotation(degrees=180)(img)
plt.imshow(np.transpose(rot, (1, 2, 0)))
plt.axis('off')

# horizontal flip
plt.subplot(2, 3, 3)
plt.title('Horizontal flip')
hflip = transforms.RandomHorizontalFlip()(img)
plt.imshow(np.transpose(hflip, (1, 2, 0)))
plt.axis('off')

# vertical flip
plt.subplot(2, 3, 4)
plt.title('Vertical flip')
vflip = transforms.RandomVerticalFlip()(img)
plt.imshow(np.transpose(vflip, (1, 2, 0)))
plt.axis('off')

# zoom
plt.subplot(2, 3, 5)
plt.title('Zoom')
zoom = transforms.RandomAffine(degrees=0, scale=(1, 1.5))(img)
plt.imshow(np.transpose(zoom, (1, 2, 0)))
plt.axis('off')

# crop
plt.subplot(2, 3, 6)
plt.title('Crop')
shift = transforms.RandomAffine(degrees=0, translate=(0.1, 0.1))(img)
plt.imshow(np.transpose(shift, (1, 2, 0)))
plt.axis('off')

plt.show()
```

Load the images into 2 datasets:
- training
- testing

```python
# create transformations to normalize the images and for data augmentation
transform = transforms.Compose([
    # convert to tensor
    transforms.ToTensor(),
    # normalize image
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5]),
    # horizontal flip
    transforms.RandomHorizontalFlip(),
    # vertical flip
    transforms.RandomVerticalFlip(),
    # zoom
    transforms.RandomAffine(degrees=0, scale=(1, 1.5)),
    # brightness
    transforms.ColorJitter(brightness=0.2, contrast=0, saturation=0, hue=0),
])

# load images from "ok_front" and "def_front" folders in grayscale and apply transformations
train_data = torchvision.datasets.ImageFolder(root=train_dir, transform=transform)
test_data = torchvision.datasets.ImageFolder(root=test_dir, transform=transform)

# create data loaders
train_loader = torch.utils.data.DataLoader(train_data, batch_size=32, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=32, shuffle=False)
```

Visualize an image from each class

```python
# plot a sample of each class
plt.figure(figsize=(12, 12))

plt.subplot(1, 2, 1)
plt.title('OK front')
ok_front = next(iter(train_loader))
ok_front = ok_front[0][ok_front[1] == 1]
plt.imshow(np.transpose(ok_front[0], (1, 2, 0)))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Defective front')
def_front = next(iter(train_loader))
def_front = def_front[0][def_front[1] == 0]
plt.imshow(np.transpose(def_front[0], (1, 2, 0)))
plt.axis('off')

plt.show()
```

#### Build the CNN model using PyTorch

```python
# build CNN model using PyTorch
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        # pooling and activitation functions
        self.pool = nn.MaxPool2d(2, 2)
        self.act1 = nn.ReLU()
        self.act2 = nn.Sigmoid()

        # convolutional layers
        # input image size is 300x300 pixels
        # conv_width = (input_width - kernel_width + 2*padding) / stride + 1
        # to preserve the spatial dimensions of the input image, set padding = ((w -1) * s - w + k) / 2
        self.conv1 = nn.Conv2d(3, 16, kernel_size=7, stride=2, padding=3)   # output size: 150x150
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)  # output size: 75x75 (pooling layer halves the size of the image)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)  # output size: 37x37 (pooling layer halves the size of the image)

        # fully connected layers
        self.fc1 = nn.Linear(64*18*18, 224)
        self.dropout = nn.Dropout(0.2)
        self.fc2 = nn.Linear(224, 1)

        # track accuracy for each epoch
        self.train_accuracy = []
        self.test_accuracy = []
        self.train_loss = []
        self.test_loss = []


    def forward(self, x):
        x = self.pool(self.act1(self.conv1(x)))
        x = self.pool(self.act1(self.conv2(x)))
        x = self.pool(self.act1(self.conv3(x)))
        x = x.view(-1, 64*18*18)
        x = self.act1(self.fc1(x))
        x = self.dropout(x)
        y = self.act2(self.fc2(x))
        return y


    def fit(self, train_loader, test_loader, n_epochs, lr=0.001, early_stop=False, stop_tol=0.001):
        # reset accuracy for each epoch
        self.train_accuracy = []
        self.test_accuracy = []

        # define loss function and optimizer
        criterion = nn.BCELoss()
        optimizer = torch.optim.Adam(self.parameters(), lr=lr)

        # train the model
        for epoch in range(n_epochs):
            for i, (imgs, lbls) in enumerate(train_loader):
                # 1. reset gradients
                optimizer.zero_grad()

                # 2. forward pass
                prob = self.forward(imgs)

                # 3. compute loss
                loss = criterion(prob, lbls.float().view(-1, 1))

                # 4. backpropagation
                loss.backward()

                # 5. update weights
                optimizer.step()

                if i % 100 == 0:
                    pred = torch.round(prob).flatten()
                    print(f'Epoch {epoch+1}/{n_epochs}, step: {i+1}/{len(train_loader)}, loss: {loss.item():.4f}, accuracy: {(pred==lbls).sum().item()/lbls.size(0)}')

            # track loss and accuracy for each epoch
            [train_accuracy, train_loss] = self.evaluate(train_loader)
            self.train_accuracy.append(train_accuracy)
            self.train_loss.append(train_loss / len(train_loader))
            [test_accuracy, test_loss] = self.evaluate(test_loader)
            self.test_accuracy.append(test_accuracy)
            self.test_loss.append(test_loss / len(test_loader))

            # print loss and accuracy for each epoch
            print(f'Epoch {epoch+1}/{n_epochs}, train loss: {train_loss / len(train_loader):.4f}, train accuracy: {self.train_accuracy[-1]:.4f}, test loss: {test_loss / len(test_loader):.4f}, test accuracy: {self.test_accuracy[-1]:.4f}\n\n')

            # early stopping
            if early_stop and epoch > 0 and self.test_loss[-2] - self.test_loss[-1] < stop_tol:
                print('Early stopping')
                break


    def evaluate(self, loader):
        correct = 0
        total = 0
        loss = 0
        criterion = nn.BCELoss()

        with torch.no_grad():
            for imgs, lbls in loader:
                pred_prob = self.forward(imgs)
                pred = torch.round(pred_prob).flatten()
                total += lbls.size(0)
                correct += (pred == lbls).sum().item()
                loss += criterion(pred_prob, lbls.view(-1, 1).float()).item()

        accuracy = correct / total

        return [accuracy, loss]

    
    def predict(self, imgs):
        with torch.no_grad():
            prob = self.forward(imgs)
            pred = torch.round(prob).flatten().tolist()

        return pred
    

    def proba(self, imgs):
        with torch.no_grad():
            prob = self.forward(imgs).flatten().tolist()

        return prob

model = CNN()
```

Test the network architecture on a single image to observe how the dimensions change through the feature extraction section.

```python
# test network architecture on a single image
data_iter = iter(train_loader)
images, labels = next(data_iter)

# architecture of the model
pool = nn.MaxPool2d(2, 2)
conv1 = nn.Conv2d(3, 16, 7, 2, 3)
conv2 = nn.Conv2d(16, 32, 3, 1, 1)
conv3 = nn.Conv2d(32, 64, 3, 1, 1)

# forward pass
print('Input:', images.shape)
x = conv1(images)
print('Conv1:', x.shape)
x = pool(x)
print('Pool1:', x.shape)
x = conv2(x)
print('Conv2:', x.shape)
x = pool(x)
print('Pool2:', x.shape)
x = conv3(x)
print('Conv3:', x.shape)
x = pool(x)
print('Pool3:', x.shape)
```

Train the model

```python
# train the model
model.fit(train_loader, test_loader, n_epochs=10, early_stop=True)
```

Make predictions and generate the confusion matrix

```python
# predict labels for all images in the test set
predictions = []
labels = []
for imgs, lbls in test_loader:
    predictions += model.predict(imgs)
    labels += lbls.tolist()

# confusion matrix
conf_matrix = confusion_matrix(labels, predictions)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Actual labels')
plt.ylabel('Predicted labels')
plt.title('Confusion matrix')
plt.show()

# classification report
print(classification_report(labels, predictions))
```

#### Plot the accuracy and loss profiles

```python
# plot accuracy
epochs = np.arange(len(model.train_accuracy)) + 1
plt.plot(epochs, model.train_accuracy, label='Train accuracy')
plt.plot(epochs, model.test_accuracy, label='Test accuracy')
plt.plot(epochs, model.train_loss, label='Train loss')
plt.plot(epochs, model.test_loss, label='Test loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.yticks(np.arange(0, 1.1, 0.1))
plt.legend()
plt.grid()
plt.show()
```
